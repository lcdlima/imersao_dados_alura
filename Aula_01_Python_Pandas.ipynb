### Desafio 05: Renomear as colunas tirando o hífen

```python
def remove_hifen(column_name):
    new_name = column_name.replace('-', '')
    dados.rename(columns={column_name: new_name}, inplace=True)

columns = list(dados.columns[5:])

for column in columns:
    remove_hifen(column)
```

Aproveitando o conhecimento que tenho da área, resolvi analisar alguns cenários que acho que fazem sentido com as informações que temos até aqui dos dados. 

```python
df_droga = dados[dados['tratamento'] == 'com_droga']
```

```python
df_droga_grouped = df_droga.groupby(['droga', 'dose', 'tempo']).mean()
df_droga_grouped.head(12)
```

```python
df_controle = dados[dados['tratamento'] == 'com_controle']
```

```python
df_controle_grouped = df_controle.groupby(['droga', 'dose', 'tempo']).mean()
df_controle_grouped.head(12)
```

#### Qual é a droga que mais aumentou a expressão do gene g0 em um menor tempo?

```python
df_droga['droga'][(df_droga['tempo'] == 24) & (df_droga['g0'] == df_droga[df_droga['tempo'] == 24]['g0'].max())]
```

#### Qual é a droga que mais diminuiu a expressão do gene g0 em um maior tempo?

```python
df_droga['droga'][(df_droga['tempo'] == 72) & (df_droga['g0'] == df_droga[df_droga['tempo'] == 72]['g0'].min())]
```
